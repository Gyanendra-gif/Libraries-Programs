```python
# 5-Python program to plot two or more lines on same plot with suitable legends of each
# line.
"""
Description:
    Function is Used to plot two or more lines on same plot with suitable legends of each
    line.
Parameter:
    Sample Axis
Return:
    None
"""
x1 = [10,20,30]
y1 = [20,40,10]
plt.plot(x1, y1, label = "line 1")  # plotting the line 1 points 
x2 = [10,20,30]
y2 = [40,10,30]
plt.plot(x2, y2, label = "line 2")  # plotting the line 2 points 
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Two or more lines on same plot with suitable legends ') # Set a title of the current axes.
plt.legend() # show a legend on the plot
plt.show() # Display a figure.
```

```python
# 6-Python program to plot two or more lines with legends, different widths and colors.
"""
Description:
    Function is Used to plot two or more lines on same plot with suitable legends of each
    line.
Parameter:
    Sample Axis
Return:
    None
"""
x1 = [10,20,30]
y1 = [20,40,10]
x2 = [10,20,30]
y2 = [40,10,30]
plt.xlabel('x - axis') # Set the x axis label of the current axis.
plt.ylabel('y - axis') # Set the y axis label of the current axis.
plt.title('Two or more lines with different widths and colors with suitable legends ')
plt.plot(x1,y1, color='blue', linewidth = 3,  label = 'line1-width-3') # Display the figure.
plt.plot(x2,y2, color='red', linewidth = 5,  label = 'line2-width-5')
plt.legend() # show a legend on the plot
plt.show()
```

```python
# 11-Python program to plot several lines with different format styles in one command using arrays.
"""
Description:
    Function is Used to plot several lines with different format styles in one command using arrays
    line.
Parameter:
    Sample Axis
Return:
    None
"""
t = np.arange(0., 5., 0.2) # Sampled time at 200ms intervals
plt.plot(t, t, 'g--', t, t**2, 'bs', t, t**3, 'r^') # green dashes, blue squares and red triangles
plt.show()
```

```python
# 12-Python program to create multiple types of charts
"""
Description:
    Function is Used to create multiple types of charts
Parameter:
    Sample Axis
Return:
    None
"""
data = [(DT.datetime.strptime('2016-10-03', "%Y-%m-%d"), 772.559998),
        (DT.datetime.strptime('2016-10-04', "%Y-%m-%d"), 776.429993),
        (DT.datetime.strptime('2016-10-05', "%Y-%m-%d"), 776.469971),
        (DT.datetime.strptime('2016-10-06', "%Y-%m-%d"), 776.859985),
        (DT.datetime.strptime('2016-10-07', "%Y-%m-%d"), 775.080017 )]

x = [date2num(date) for (date, value) in data]
y = [value for (date, value) in data]
fig = plt.figure()
graph = fig.add_subplot(111)
graph.plot(x,y,'r-o') # Plot the data as a red line with round markers
graph.set_xticks(x) # Set the xtick locations

# Set the xtick labels
graph.set_xticklabels(
        [date.strftime("%Y-%m-%d") for (date, value) in data]
        )
plt.show()
```

```python
# 13-Python program to display the grid and draw line charts of the closing value of
# Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with
# linestyle -, width .5. and color blue.
"""
Description:
    Function is Used to display the grid and draw line charts of the closing value
Parameter:
    Sample Axis
Return:
    None
"""
data = [(DT.datetime.strptime('2016-10-03', "%Y-%m-%d"), 772.559998),
        (DT.datetime.strptime('2016-10-04', "%Y-%m-%d"), 776.429993),
        (DT.datetime.strptime('2016-10-05', "%Y-%m-%d"), 776.469971),
        (DT.datetime.strptime('2016-10-06', "%Y-%m-%d"), 776.859985),
        (DT.datetime.strptime('2016-10-07', "%Y-%m-%d"), 775.080017 )]

x = [date2num(date) for (date, value) in data]
y = [value for (date, value) in data]

fig = plt.figure()

graph = fig.add_subplot(111)

graph.plot(x,y,'r-o') # Plot the data as a red line with round markers

graph.set_xticks(x) # Set the xtick locations

# Set the xtick labels
graph.set_xticklabels(
        [date.strftime("%Y-%m-%d") for (date, value) in data]
        )

plt.xlabel('Date') # naming the x axis
plt.ylabel('Closing Value') # naming the y axis
plt.title('Closing stock value of Alphabet Inc.') # giving a title  
plt.grid(linestyle='-', linewidth='0.5', color='blue') # Customize the grid
plt.show()
```

```python
# 14-Python program to display the grid and draw line charts of the closing value of
# Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with
# rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn
# off ticks.
"""
Description:
    Function is Used to display the grid and draw line charts of the closing value Customized the grid lines with
    rendering with a larger grid (major grid) and a smaller grid (minor grid)
Parameter:
    Sample Axis
Return:
    None
"""
data = [(DT.datetime.strptime('2016-10-03', "%Y-%m-%d"), 772.559998),
        (DT.datetime.strptime('2016-10-04', "%Y-%m-%d"), 776.429993),
        (DT.datetime.strptime('2016-10-05', "%Y-%m-%d"), 776.469971),
        (DT.datetime.strptime('2016-10-06', "%Y-%m-%d"), 776.859985),
        (DT.datetime.strptime('2016-10-07', "%Y-%m-%d"), 775.080017 )]

x = [date2num(date) for (date, value) in data]
y = [value for (date, value) in data]

fig = plt.figure()

graph = fig.add_subplot(111)
graph.plot(x,y,'r-o') # Plot the data as a red line with round markers

graph.set_xticks(x) # Set the xtick locations

# Set the xtick labels
graph.set_xticklabels(
        [date.strftime("%Y-%m-%d") for (date, value) in data]
        )

plt.xlabel('Date') # naming the x axis
plt.ylabel('Closing Value') # naming the y axis
plt.title('Closing stock value of Alphabet Inc.') 
plt.grid(linestyle='-', linewidth='0.5', color='blue') # Customize the grid
plt.show()
```
