```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# Write a program to draw a violin plot of sex against total_bill for a given dataset
"""
Description:
    Function is Used to draw a violin plot of sex against total_bill for a given dataset
Parameter:
    Sample Array
Return:
    None
"""
f, ax = plt.subplots(figsize=(6,6))
sns.set_theme(style="whitegrid")
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True,linewidth=1,palette='rocket')
sns.despine(left=True,bottom=True)
```
