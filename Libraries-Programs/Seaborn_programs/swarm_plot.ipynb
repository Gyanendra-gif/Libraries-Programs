```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# program to draw a swarm plot of total bill against size for a given dataset
"""
Description:
    Function is Used to draw a swarm plot of total bill against size for a given dataset
Parameter:
    Sample Array
Return:
    None
"""
tips = sns.load_dataset('tips')
sns.swarmplot(x='size',y='total_bill',data=tips, palette="rocket",hue='sex')
sns.despine(offset=10)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips, palette="rocket",hue='sex')
sns.despine(offset=10)
```
